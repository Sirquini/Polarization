# Formalization of static elements of the model: Introduction

*Static elements* of the model are those which simply describe a "still snapshot" of the world, with all the information necessary to evaluate how "polarized" the scenario is wrt. to some apporpriate measure of polarization.

-  A (finite) set $\mathcal{A} = \{a_0, a_1, \ldots, a_{n-1} \}$ of $n \geq 1$ *agents*.

# Formalization of static elements of the model: Polarization Measures
We consider the following polarization measures.

## The Esteban-Ray polarizatoin measure

This measure was proposed in [[Esteban & Ray, 1994]](https://www.econ.nyu.edu/user/debraj/Courses/Readings/Esteban%20Ray94.pdf). 

Consider a (finite) discret set $\mathcal{Y} = \{y_0,y_1,\ldots,y_{k-1}\}$ of size $k$.
Let $(\pi,y) = (\pi_0,\pi_1,\ldots,\pi_{k-1},y_0,y_1,\ldots,y_{k-1})$ be a *distribution* on $\mathcal{Y}$ st. $\pi_i$ is the frequency of value $y_i \in \mathcal{Y}$ in the distribution. Wlog. we can assume the values of $\pi_i$ are all non-zero and add up to 1 (as in a real distribution). 
The *Esteban-Ray polarization measure* is defined as

$$
\mathit{Pol}_{\mathit{ER}}(\pi,y) = K \sum_{i=0}^{k-1}\sum_{j=0}^{k-1}\pi_i^{1+\alpha}\pi_j\mid y_i - y_j\mid, 
$$

where $K>0$ is a constant, and typically $\alpha \approx 1.6$.
   
### Converting a belief state into the distribution format used by the Esteban-Ray measure

Note that $\mathit{Pol}_{\mathit{ER}}$ is defined on a *discret distribution*, whereas in our model a general polarization metric is defined on a belief state $B_p : \mathcal{A} \rightarrow [0,1]$. 
To properly apply $\mathit{Pol}_{\mathit{ER}}$ to our setup we need to convert the belief state $B_p$ into an appropriate distribution $(\pi,y)$ as expected by the measure.
We can do it as follows.
   
Given a number $k > 0$ of *(discretization) bins*, define $\mathcal{Y} = \{y_1,y_2,\ldots,y_k\}$ as the discretized version of the belief strenght $[0,1]$ in such a way that each $y_i$ represents all beliefs with strenght in range
   
$$
y_i = \left[\frac{i}{k},\frac{i+1}{k}\right],
$$
   
and define the weight $\pi_i$ of value $y_i$ as
   
$$
\pi_i = \frac{\sum_{a \in \mathcal{A}} \mathit{Ind}(B_p(a) \in y_i)}{\mid \mathcal{A} \mid},
$$
   
where $\mathit{Ind}$ is the *indicator function* returning 1 if its parameter is true, and 0 otherwise.

# Formalization of dynamic elements of the model: Introduction

*Dynamic elements* of the model are those which describe how a belief state (or state of belief) evolves over time as agents interact by exchanging messages and updating their beliefs.

We define:

-  An *influence graph* $I : \mathcal{A} \times \mathcal{A} \rightarrow [0,1]$ s.t. $I(a_i,a_j)$ represents the influence of agent $a_i$ on agent $a_j$, for all $a_i, a_j \in \mathcal{A}$. The higher this value, the stronger the influence exerted.

-  A *time frame* $\mathcal{T} = \{0, 1, 2, \ldots, t_{\mathit{max}}\}$ representing the discrete passing of time.
   Each $t \in \mathcal{T}$ is called a *time step*.

-  A *family of belief states* $\{B_{p}^{t}\}_{t \in \mathcal{T}}$ parametrized by the time frame, in which each $B_{p}^{t}$ is the belief state wrt. proposition $p$ at time step $t \in \mathcal{T}$.

-  A *(belief) update function* $U_{p} : B_p \times I \rightarrow B_p$. The evolution of the model is captured as $B_p^{t+1} = U_{p}(B_p^t, I)$ is the update of agents' beliefs from time step $t$ to time step $t+1$.


There are several ways of defining update functions, we will discuss some in more detail below.


# Formalization of dynamic elements of the model: Update Functions

We consider the following (belief) update functions:

## Classic update (Sophia's original proposal)

**Intuition:** This is similar to classic Bayesian update. 
The impact of agent $a_j$'s belief on agent $a_i$'s  is proportinal to the influence $I(a_j,a_i)$ of $a_j$ on $a_i$, and to the difference $B_p^t(a_j) - B_p^t(a_i)$ in their beliefs.
   
The *classic update* of agent $a_i$'s belief given the interaction with agent $a_j$ at time $t$ is formalized as
   
$$
B_p^{t+1}(a_i \mid a_j) = B_p^t(a_i) + I(a_j,a_i) \left( B_p^t(a_j) - B_p^t(a_i) \right).
$$
   
The *overall classic update* of agent $a_i$'s belief given the interaction with all other agents in $\mathcal{A}$ at time $t$ is formalized as:
   
$$
B_p^{t+1}(a_i) = \frac{1}{\mid\mathcal{A}\mid} \sum_{a_j \in \mathcal{A}} B_p^{t+1}(a_i \mid a_j).
$$

## Confirmation bias update

**Intuition:** Here we consider that people tend to give more weight to evidence that already supports their views than to evidence that contradicts their views.
More precisely, when agent $a_j$ presents agent $a_i$ with evidence for or against proposition $p$, the update function will still move an agent $a_i$'s belief in the direction of the belief of agent $a_j$'s, proportionally to the influence $I(a_j,a_i)$ that $a_j$ has over $a_i$, but with the following caveat: the update will be stronger when $a_j$'s belief is similar to $a_i$'s than when it's dissimilar.

The *confirmation-bias update* of agent $a_i$'s belief given the interaction with agent $a_j$ at time $t$ is formalized as
   
$$
B_p^{t+1}(a_i \mid a_j) = B_p^t(a_i) + \mathit{f_\mathit{confbias}} * I(a_j,a_i) \left( B_p^t(a_j) - B_p^t(a_i) \right),
$$
   
where $\mathit{f_\mathit{confbias}}$ is a *confirmation-bias factor*.
   
We consider at least two definitions of the confirmation-bias factor:
   
-  **smooth bias:** proportional to the difference of belief level:
   
$$
\mathit{f_\mathit{confbias}} = 1 - \mid B_p^t(a_j) - B_p^t(a_i) \mid,
$$
   
-  **sharp bias:** a simple discount factor of 0.5 when the beliefs of each agent are not both at the same time mostly positive ($\geq 0.5$) or at the same time mostlly negative ($< 0.5$):
   
$$
\mathit{f_\mathit{confbias}} = 
\begin{cases} 
    1, & \text{if $(B_p^t(a_j)- 0.5) * (B_p^t(a_i) - 0.5) \geq 0$, }\\
    0.5 & \text{otherwise.}
\end{cases}
$$
   
Again, the *overall confirmation-bias update* of agent $a_i$'s belief given the interaction with all other agents in $\mathcal{A}$ at time $t$ is formalized as:   

$$
B_p^{t+1}(a_i) = \frac{1}{\mid\mathcal{A}\mid} \sum_{a_j \in \mathcal{A}} B_p^{t+1}(a_i \mid a_j).
$$
   
## Backfire effect update

**Intuition:** "The effect is claimed to be that when, in the face of contradictory evidence, established beliefs do not change but actually get stronger. The effect has been demonstrated experimentally in some psychological tests, where subjects are given data that either reinforces or goes against their existing biases â€” and in most cases people can be shown to increase their confidence in their prior position regardless of the evidence they were faced with. In a pessimistic sense, this would make most refutations useless." [Source: https://rationalwiki.org/wiki/Backfire_effect]
   
"One hypothesis as to why this and the backfire effect happens is that you spend much more time considering information you disagree with than you do information you accept. 
(...) 
Information which lines up with what you already believe passes through the mind like a vapor, but when you come across something which threatens your beliefs, something which conflicts with your preconceived notions of how the world works, you seize up and take notice. Some psychologists speculate there is an evolutionary explanation. Your ancestors paid more attention and spent more time thinking about negative stimuli than positive because bad things required a response. Those who failed to address negative stimuli failed to keep breathing."
   [Source:  https://youarenotsosmart.com/2011/06/10/the-backfire-effect/] 
   
To define an update function considering the backfire effect, we start with the following.

-  Define the *belief gap* between agents $a_i$ and $a_j$ at time $t$ as $\delta_{p}^{t}(i,j) = \left| B_p^{t}(a_i) - B_p^{t}(a_j) \right|$.

-  Define a *backfire belief threshold* as a value $\mathit{backfire\_belief\_threshold} \in [0,1]$.

-  Define a *backfire influence threshold* as a value $\mathit{backfire\_influence\_threshold} \in [0,1]$.


When an agent $a_j$ is interacting with an agent $a_i$, the bacfire effect is triggered when 2 conditions are satisfied:

1. the agent's beliefs is too different, i.e., if 
$$
\delta_{p}^{t}(i,j) \geq \mathit{backfire\_belief\_threshold},
$$
   
and
   
2. the agent $a_j$'s influence on agent $a_i$ is above a minimum, i.e., if

$$
I(a_j, a_i) \geq \mathit{backfire\_influence\_threshold}.
$$

If the backfire effect is trigered, agent $a_i$ won't consider $a_j$'s message in the regular way (i.e., by letting $a_j$'s belief influence $a_i$'s belief in a direct way), but rather $a_i$'s belief will get stronger in its original direction exactly because $a_j$ disagrees with $a_i$.
   
To model that, we need an update function from belief value to belief value mapping $B_p^{t}(a_i) \mapsto B_p^{t+1}(a_i)$ that **doesn't weaken $a_i$'s original belief**, ie.,:

1. if $B_p^{t}(a_i) \geq 0.5$ then $B_p^{t+1}(a_i) \geq B_p^{t}(a_i)$, and 
2. if $B_p^{t}(a_i) \leq 0.5$ then $B_p^{t+1}(a_i) \leq B_p^{t}(a_i)$.
   
These points are captured by the figure below.
   
![image](figures/backfire-motivation.png)
   
The above suggests the backfire update is a form of *step function*, which can be studied here [https://en.wikipedia.org/wiki/Heaviside_step_function#Analytic_approximations] for more details. 
But the shape of the function is:

$$
B_{p}^{t+1}(a_i) = \frac{1}{1 + e^{-2 \cdot \mathit{backfire\_factor} \cdot z}},
$$

where $z = 2 \cdot B_p^{t} - 1$ is a transformation to bring the step-curve from the interval $[-1,1]$ to $[0,1]$, and $\mathit{backfire\_factor}$ is a parameter controlling how "steepy" is the step (ie., the greter the parameter the closer the curve is to a sharp step).
The parameter $\mathit{backfire\_factor}$ serves our purpose of having different levels of strenght for the backfire effect: the steeper the graph, the stronger the effect.
   
We also want the strengthneing of $a_i$'s belief to be:
   
3. more pronounced if the *belief gap* $\delta_{p}^{t}(i,j)$ is high.
   (The intuition here is that the more the person disagrees with you, the more you will fight the evidence they show to you.)
      
4. less pronounced if the influence $I(a_j,a_i)$ of agent $a_j$ on agent $a_i$ is low.     
   (The intuition here is that you more easily dismiss evidence provided by people you don't respect much.)
      
The above suggests that the factor controlling how steep the curve is can be:
      
$$
\mathit{backfire\_factor} = \frac{\delta_{p}^{t}(i,j)}{I(a_j,a_i)}
$$

Finally, we put it all together to get our definition of backfire effect update function.

The *backfire update* of agent $a_i$'s belief given the interaction with agent $a_j$ at time $t$ is formalized as

$$
B_p^{t+1}(a_i \mid a_j) = 
\begin{cases}
B_p^t(a_i) + I(a_j,a_i) \left( B_p^t(a_j) - B_p^t(a_i) \right), & \text{if $\delta_{p}^{t}(i,j) < \mathit{backfire\_belief\_threshold}$ or $I(a_j,a_i) < \mathit{backfire\_influence\_threshold}$} \\
\displaystyle \min\left(B_p^{t}(a_i),\frac{1}{1+e^{-2 \cdot \mathit{backfire\_factor} \cdot z}}\right), & \text{if $\delta_{p}^{t}(i,j) \geq \mathit{backfire\_belief\_threshold}$ and $I(a_j,a_i) \geq \mathit{backfire\_influence\_threshold}$ and $B_p^{t}(a_i) \leq 0.5$} \\
\displaystyle \max\left(B_p^{t}(a_i),\frac{1}{1+e^{-2 \cdot \mathit{backfire\_factor} \cdot z}}\right), & \text{if $\delta_{p}^{t}(i,j) \geq \mathit{backfire\_belief\_threshold}$ and $I(a_j,a_i) < \mathit{backfire\_influence\_threshold}$ and $B_p^{t}(a_i) \geq 0.5$} 
\end{cases}
$$

where $z = 2\cdot B_p^{t}(a_i) - 1$, and $\mathit{backfire\_factor} = \frac{\delta_{p}^{t}(i,j)}{I(a_j,a_i)}$.
   
The *overall classic update* of agent $a_i$'s belief given the interaction with all other agents in $\mathcal{A}$ at time $t$ is formalized as:
   
$$
B_p^{t+1}(a_i) = \frac{1}{\mid\mathcal{A}\mid} \sum_{a_j \in \mathcal{A}} B_p^{t+1}(a_i \mid a_j).
$$

# Implementation: Representing belief states

A belief state (or belief state) $B_p$ is represented as a vector of flots **belief_vec**, in which **belif_vec[i]** = $B_p(a_i)$ for each $0 \leq i \leq n-1$.

In our initial simulations we are considering that the size of the set $\mathcal{A}$ of agents is $n=10$ or $n=4$, and four configurations of initial belief states:

-  A *"uniform"* belief state **belief_vec_uniform**, in which each one of the agents has a different level of belief on $p$.
-  A *"mildly polarized"* belief state **belief_vec_mild**, in which half of the agents have belief around level 0.23, and half around level 0.77.
-  An *"extremely polarized"* belief state **belief_vec_extreme**, in which half of the agents has minimum belief (0), and half has maximum (1).
-  A *"3-pole"* belief state **belief_vec_triple**, in which 3 agents have belief at level 0.0, 4 agents have belief at level 0.5, and 3 agents have belief at level 1.
-  A *"consesus"* belief state **belief_vec_consensus**, in which all agents have the same level of belief (0.5) on $p$.

```python
## Builds uniform belief state.
def Build_Belief_Vec_Uniform(num_agents):
    belief_vec = [0] * num_agents
    for i in range(num_agents):
        belief_vec[i] = i/(num_agents - 1)
    return belief_vec

belief_vec_uniform = Build_Belief_Vec_Uniform(num_agents)

## Builds mildly polarized belief state, in which 
## half of agents has belief decreasing from 0.25, and
## half has belief increasing from 0.75, all by the given step.
def Build_Belief_Vec_Mild(num_agents, low_pole, high_pole, step):
    belief_vec = [0] * num_agents
    middle = math.ceil(num_agents/2)
    for i in range(0,middle):
        belief_vec[i] = max(low_pole - step*(middle - i - 1), 0)
    for i in range(middle,num_agents):
        belief_vec[i] = min(high_pole - step*(middle - i), 1)
    return(belief_vec)
    
belief_vec_mild = Build_Belief_Vec_Mild(num_agents, low_pole, high_pole, belief_step)

## Builds extreme polarized belief state, in which half
## of the agents has belief 0, and half has belief 1.
def Build_Belief_Vec_Extreme(num_agents):
    belief_vec = [0] * num_agents
    middle = math.ceil(num_agents/2)
    for i in range(middle,num_agents):
        belief_vec[i] = 1
    return belief_vec

belief_vec_extreme = Build_Belief_Vec_Extreme(num_agents)

## Builds three-pole belief state, in which each 
## one third of the agents has belief 0, one third has belief 0.4,
## and one third has belief 1.
def Build_Belief_Vec_Triple(num_agents):
    belief_vec = [0] * num_agents
    one_third = math.floor(1/3 * num_agents)
    two_thirds = math.ceil(2/3 * num_agents)
    for i in range(one_third, two_thirds):
        belief_vec[i] = 0.5
    for i in range(two_thirds, num_agents):
        belief_vec[i] = 1
    return belief_vec

belief_vec_triple = Build_Belief_Vec_Triple(num_agents)

## Builds consensus belief state, in which each 
## all agents have same belief.
def Build_Belief_Vec_Consensus(num_agents, belief):
    belief_vec = [belief] * num_agents
    return belief_vec

belief_vec_consensus = Build_Belief_Vec_Consensus(num_agents, consensus_value)
```

# Implementation: The Esteban-Ray polarization measure

Below we implement the Esteban-Ray polarization measures.

First we present a function **Belief_2_Dist** to convert a belief state $B_p$, represented as a vector **belief_vec** into a discrete distribution $(\pi,y)$, represented as a matrix **dist** such that:

-  The first row of **dist** contains the description of each discretized interval.
   More precisely, each 
   
   $$
   \text{dist[0,i]} = \frac{(i+0.5)}{k}
   $$ 
   
   corresponds to the middle point of the interval $y_i = \left[\frac{i}{k},\frac{i+1}{k}\right]$.
   
-  The second row of **dist** contains the proportions of agents whose beliefs fall into each discretized interval.
   More precisely, each 
   
   $$
   \text{dist}[1,i] = \pi_i = \frac{\sum_{a \in \mathcal{A}} \mathit{Ind}(B_p(a) \in y_i)}{\mid \mathcal{A} \mid}.
   $$ 
   
Second, we implement a function **Pol_ER** to compute the Esteban-Ray polarization of a distribution.

```python
## This function takes a belief state belief_vec and discretizes it into a number num_bins of bins.
def Belief_2_Dist(belief_vec,num_bins):
    # stores labels of bins
    bin_labels = [0] * num_bins
    # the value of a bin is the medium point of that bin
    for i in range(0,num_bins):
        bin_labels[i] = (i+0.5)/num_bins
    # stores the distribution of labels
    bin_prob = [0] * num_bins
    # for all agents...
    for i in range(0,len(belief_vec)):
        # computes the bin into which the agent's belief falls
        bin = math.floor(belief_vec[i]*num_bins)
        # treats the extreme case in which belief is 1, putting the result in the right bin.
        if bin == num_bins:
            bin = num_bins - 1
        # updates the frequency of that particular belief
        bin_prob[bin] += 1/len(belief_vec)
    # bundles into a matrix the bin_labels and bin_probabilities.
    dist = np.array([bin_labels,bin_prob])
    # returns the distribution.
    return dist

## Computes the Esteban-Ray polarization of a distribution.
def Pol_ER(dist):
    # recover bin labels
    bin_labels = dist[0]
    # recover bin probabilities
    bin_prob = dist[1]
    # number of bins
    num_bins = len(bin_labels)
    # initialize polarization measure
    pol = 0
    # computes the esteban-ray measure
    for i in range(0,num_bins):
        for j in range(0,num_bins):
            pol += ( (bin_prob[i]**(1+alpha)) * bin_prob[j] ) * abs(bin_labels[i] - bin_labels[j]) 
    # scales the measure by the constant K, and returns it.
    return K*pol

## This function is for the visualization of distributions.
def Plot_Belief_Polarization(belief_vec, file_name='no-name', color='b', step = -1, pol_value=-1):
    
    ## Plots graph
    plt.figure(dpi=histogram_resolution, figsize=(histogram_size_x, histogram_size_y))
    
    ## Rounds belief_vec to desired precision.
    belief_vec_rounded = np.round(belief_vec, decimals = 3)
    
    title = ''
    if title_on_histogram:
        title += 'Histogram of Polarization\n'        
            
    if step != -1:
        title += 'Step: ' + str(step)
    
    if step != -1 and pol_value != -1:
        title += '  -  '
        
    if pol_value != -1:
        title += 'Esteban-Ray value: ' + str("%.3f" % pol_value)
    
    plt.title(title)
    xlabel = 'Belief state: \n ['
    for i in range(len(belief_vec_rounded)-1):
        xlabel += "%.3f, " % belief_vec_rounded[i]
    xlabel += "%.3f]" % belief_vec_rounded[len(belief_vec_rounded)-1]
    plt.xlabel(xlabel)
    plt.ylabel('Proportion \n n = ' + str(len(belief_vec_rounded)))
    
    ## Normalizes y-axis to percentage
    plt.ylim(0,1)
    normalized_2_percenage = np.ones(len(belief_vec_rounded))/len(belief_vec_rounded)
    
    ## Plots graph
    plt.hist(belief_vec_rounded, bins=num_bins, range=(0,1), histtype='stepfilled', facecolor=color, weights=normalized_2_percenage)
    
    ## Creates figure name, removing white spaces
    fig_name = fig_folder + file_name + fig_extension
    fig_name = fig_name.replace(" ","")
    
    if file_name != 'no-name':
        plt.savefig(fig_name, bbox_inches='tight')
    
    plt.show()    
    plt.close()
    
# def Print_Step(t, belief_vec, pol_value):
# print('Belief at t =', "%2d" % t, ':', "%.3f "*len(belief_vec) % tuple(belief_vec))
# print('Polarization : ', "%0.3f" % pol_value, '\n')
```

# Implementation: Representing influence graphs

An *influence graph* $I : \mathcal{A} \times \mathcal{A} \rightarrow [0,1]$ is represented as a $\mid \mathcal{A} \mid \times \mid \mathcal{A} \mid$ matrix **int_graph** (for influence graph) in which each **int_graph[i,j] = I(a_i,a_j)**  is the influence of agent $a_i$ on agent $a_j$.

In our initial simulations we are considering that the size of the set $\mathcal{A}$ of agents is $n=10$, and five configurations of influence graphs:

-  A *"clique"* influence graph **int_graph_clique** in which every agent has a moderate influence on every agent. 
   This represents a world in which interacts with everyone, like in an idealized totally connected social network.
-  An *"disconnected"* influence graph **int_graph_2_groups_disconnected** in which agents are divided into two groups in such a way that people within a same group interact closely, but people from different groups do not interact at all.
   This represets a world in which society is clearly divided into two "sub-societies" that evolve separately with no communication what so ever.
-  A *"faintly-connected"* influence graph **int_graph_2_groups_faint** in which agents are divided into two groups in such a way that people within a same group interact closely, but people from different groups interact only faintly .
   This represets a world in which society is clearly divided into two "sub-societies" that evolve separately, with only faint communication between them.
-  A *"balanced 2-influencers"* influence graph **int_graph_2_influencers_balanced** in which there are two agents that higlhly influence all others, but are barely influentiable by anyone else. In this case both influences have comparable influence in everyone.
    This could represent, for instance, a world with 2 very realiable TV networks with similar levels of influence interact with the whole of society, but are only moderately influenced by this very society (eg., by chekcing ratings and choosing what news to cover accordingly). Moreover, in this scenario we assume that people interact with each other, but with way less influence than the TV channels have.
-  An *"unbalanced 2-influencers"* influence graph **int_graph_2_influencers_unbalanced** in which there are two agents that higlhly influence all others, but are not influentiable by anyone. In this case one agent has much higher influence than the other.
    This could represent, for instance, a world with 2 very realiable TV networks with dissimilar levels of influence interact with the whole of society, but are only moderately influenced by this very society (eg., by chekcing ratings and choosing what news to cover accordingly). Moreover, in this scenario we assume that people interact with each other, but with way less influence than the TV channels have.

```python
## Influence graph for "clique" scenario.
def Build_Int_Graph_Clique(num_agents, 
                          clique_belief_value):
    int_graph = np.array(np.full((num_agents,num_agents), clique_belief_value))
    return(int_graph)

int_graph_clique= Build_Int_Graph_Clique(num_agents, 
                                       clique_belief_value)

## Influence graph for "disconnected" scenario.
def Build_Int_Graph_2_Groups_Disconnected(num_agents, 
                                          groups_disconnected_belief_value):
    int_graph = np.array(np.zeros((num_agents,num_agents)))
    middle = math.ceil(num_agents/2)
    for i in range(0,middle):
        for j in range(0,middle):
            int_graph[i,j] = groups_disconnected_belief_value
    for i in range(middle,num_agents):
        for j in range(middle,num_agents):
            int_graph[i,j] = groups_disconnected_belief_value
    return(int_graph)   

int_graph_2_groups_disconnected = Build_Int_Graph_2_Groups_Disconnected(num_agents, 
                                                                        groups_disconnected_belief_value)

## Influence graph for "faintly-connected" scenario.
def Build_Int_Graph_2_Groups_Faint(num_agents, 
                                   groups_faintly_belief_value_weak, 
                                   groups_faintly_belief_value_strong):
    int_graph = np.array(np.full((num_agents,num_agents), groups_faintly_belief_value_weak))
    middle = math.ceil(num_agents/2)
    for i in range(0,middle):
        for j in range(0,middle):
            int_graph[i,j] = groups_faintly_belief_value_strong
    for i in range(middle,num_agents):
        for j in range(middle,num_agents):
            int_graph[i,j] = groups_faintly_belief_value_strong
    return(int_graph)   

int_graph_2_groups_faint = Build_Int_Graph_2_Groups_Faint(num_agents, 
                                                          groups_faintly_belief_value_weak, 
                                                          groups_faintly_belief_value_strong)

## Influence graph for "balanced 2-influencers" scenario.
def Build_Int_Graph_2_Influencers_Balanced(num_agents, 
                                           influencers_balanced_incoming_both, 
                                           influencers_balanced_outgoing_both, 
                                           influencers_balanced_others):
    int_graph = np.array(np.full((num_agents,num_agents), influencers_balanced_others))
    ## Sets the influence of agent 0 on all others
    for i in range(0,num_agents-1):
        int_graph[0,i] = influencers_balanced_outgoing_both        
    ## Sets the influence of agent n-1 on all others
    for i in range(1,num_agents):
        int_graph[num_agents-1,i] = influencers_balanced_outgoing_both
    ## Sets the influence of all other agents on agent 0.
    for i in range(1,num_agents):
        int_graph[i,0] = influencers_balanced_incoming_both
    ## Sets the influence of all other agents on agent n-1.
    for i in range(0,num_agents-1):
        int_graph[i,num_agents-1] = influencers_balanced_incoming_both    
    return(int_graph)

int_graph_2_influencers_balanced = Build_Int_Graph_2_Influencers_Balanced(num_agents, 
                                                                          influencers_balanced_incoming_both, 
                                                                          influencers_balanced_outgoing_both, 
                                                                          influencers_balanced_others)

## Influence graph for "unbalanced 2-influencers" scenario
def Build_Int_Graph_2_Influencers_Unalanced(num_agents, 
                                            influencers_unbalanced_outgoing_first, 
                                            influencers_unbalanced_outgoing_second,
                                            influencers_unbalanced_incoming_first,
                                            influencers_unbalanced_incoming_second,
                                            influencers_unbalanced_others):    
    int_graph = np.array(np.full((num_agents,num_agents), influencers_unbalanced_others))
    ## Sets the influence of agent 0 on all others
    for i in range(0,num_agents-1):
        int_graph[0,i] = influencers_unbalanced_outgoing_first       
    ## Sets the influence of agent n-1 on all others
    for i in range(1,num_agents):
        int_graph[num_agents-1,i] = influencers_unbalanced_outgoing_second
    ## Sets the influence of all other agents on agent 0.
    for i in range(1,num_agents):
        int_graph[i,0] = influencers_unbalanced_incoming_first
    ## Sets the influence of all other agents on agent n-1.
    for i in range(0,num_agents-1):
        int_graph[i,num_agents-1] = influencers_unbalanced_incoming_second  
    return(int_graph)

int_graph_2_influencers_unbalanced = Build_Int_Graph_2_Influencers_Unalanced(num_agents,
                                                                             influencers_unbalanced_outgoing_first, 
                                                                             influencers_unbalanced_outgoing_second,
                                                                             influencers_unbalanced_incoming_first,
                                                                             influencers_unbalanced_incoming_second,
                                                                             influencers_unbalanced_others)
```

# Implementation: Update functions

Here we implement the update function **Update_All** that takes:

-  a belief state **belief_vec**
-  an influence graph **int_graph**
-  a string **update_type** indicating the type of update, namely
   + 'classic' for classic bayesian
   + 'confbias_smooth' for the one that models confirmation bias in a smooth way
   + 'confbias_sharp' for the one that models confirmation bias in a sharp way
   + 'backfire' for the one that that models backfire effect
-  a float **confbias_discount** indicating the confirmation bias discount
-  a float **backfire_belief_threshold** indicating the backfire threshold on beliefs
-  a float **backfire_influence_threshold** indicating the backfire threshold on influences

and returns a new belief state that is the update of the first one.
The function model the passing of one time step.


```python
## Updates the whole belief state.
def Update_All(belief_vec, int_graph, update_type, confbias_discount, backfire_belief_threshold, backfire_influence_threshold):
    num_agents = len(belief_vec)
    belief_vec_new = [0] * num_agents
    for agent_i in range(0,num_agents):
        belief_vec_new[agent_i] = Update_Allx1(belief_vec, int_graph, agent_i, update_type, confbias_discount, backfire_belief_threshold, backfire_influence_threshold)
    return (belief_vec_new)

## Updates the belief value of one individual ai considering the effect of all other agents on him.
def Update_Allx1(belief_vec, int_graph, agent_i, update_type, confbias_discount, backfire_belief_threshold, backfire_influence_threshold):
    belief_ai = 0
    num_agents = len(belief_vec)
    for agent_j in range(0,num_agents):
        belief_ai += Update_1x1(belief_vec[agent_i],belief_vec[agent_j],int_graph[agent_j,agent_i], update_type, confbias_discount, backfire_belief_threshold, backfire_influence_threshold)
    return belief_ai / num_agents

## Updates the belief value of one individual ai considering the effect one other individual aj on him.
def Update_1x1(belief_ai, belief_aj, influence, update_type, confbias_discount, backfire_belief_threshold, backfire_influence_threshold):
    # if the update type is classic...
    if update_type == 'classic':
        return belief_ai + influence * (belief_aj - belief_ai)
    # if the update type considers confirmation bias...
    elif update_type == 'confbias_smooth':
        confbias_factor = 1 - np.abs(belief_aj - belief_ai) 
        return belief_ai + confbias_factor * influence * (belief_aj - belief_ai)
    elif update_type == 'confbias_sharp':
        if ( (belief_aj - 0.5) * (belief_ai - 0.5) >= 0 ):
            confbias_factor = 1
        else:
            confbias_factor = confbias_discount
        return belief_ai + confbias_factor * influence * (belief_aj - belief_ai)
    # if the update type considers the backfire effect...
    elif update_type == 'backfire':
        ## computes the absolute difference in belief between agents
        deltaij = np.abs(belief_ai-belief_aj)
        
        ## if backfire is to be triggered...
        if deltaij >= backfire_belief_threshold and influence >= backfire_influence_threshold:
            
            ## computes the backfire factor
            backfire_factor = deltaij/influence
        
            ## This transformation takes the step function from (-1,1) to (0,1)
            z = (2*belief_ai)-1
                
            ## Computes the backfire update
            new_belief = 1/(1+math.e**(-2*backfire_factor*z))
            if belief_ai <= 0.5:
                new_belief = min(belief_ai, new_belief)
            else:
                new_belief = max(belief_ai, new_belief)
        else:
            new_belief = belief_ai + influence * (belief_aj - belief_ai)
            
        return new_belief
    # if the update type is not defined...
    else: 
        return 0
```

# Simulation: The function that runs multiple experiments

The function **run_simulation** runs a simulation taking as parameters:

-  an initinal belief state **belief_vec**
-  an influence graph **int_graph**
-  the maximum number **max_time** of time steps allowed
-  the number **num_bins** of bins used in the discretization of a belief state into a distribution used for the Esteban-Ray polarization measure
-  the type **update_type**, which can be:
   + 'classic'
   + 'confbias_smooth'
   + 'confbias_sharp'
   + 'backfire' 
-  the confirmation bias factor **confbias_factor** if necessary
-  the backfire factor **backfire_belief_threshold**
-  the backfire factor **backfire_influence_threshold**
-  the printing mode **print_mode**, which can be:
   +  'all' : prints all intermediate belief vectors
   +  'last' : prints the only the last (final) belief vector
   +  'none' : prints none of the belief vectors

```python
def run_simulation(belief_vec, 
                   int_graph, 
                   max_time, 
                   num_bins, 
                   update_type, 
                   confbias_discount, 
                   backfire_belief_threshold,
                   backfire_influence_threshold,
                   print_mode):

    ## Creates temporary values to store evolution with time.
    belief_vec_temp = belief_vec
    dist = Belief_2_Dist(belief_vec_temp,num_bins)
    pol_temp = Pol_ER(dist)
    
    ## Prints initial state of simulation, if demanded.
    if print_mode == 'all' or print_mode == 'last':
        #Print_Step(0, belief_vec_temp, pol_temp)
        Plot_Belief_Polarization(belief_vec_temp, color='b', step=0, pol_value=pol_temp)
    
    pol_hist = np.array([pol_temp])

    ## Executes simulation for max_time steps.
    for t in range(1,max_time):
        ## Updates belief vector
        belief_vec_temp = Update_All(belief_vec_temp, int_graph, update_type, confbias_discount, backfire_belief_threshold, backfire_influence_threshold)
        ## Discretize belief state as necessary for computing Esteban-Ray polarization.
        dist = Belief_2_Dist(belief_vec_temp,num_bins)
        ## Computs Esteba-Ray polarization.
        pol_temp = Pol_ER(dist)
        ## Appends new polarization level to history log.
        pol_hist = np.append(pol_hist,[pol_temp])
        ## Prints results, if demanded.
        if print_mode == 'all':
            #Print_Step(t, belief_vec_temp, pol_temp)
            Plot_Belief_Polarization(belief_vec_temp, color='b', step=t, pol_value=pol_temp)
    
    ## Prints results, if demanded.
    if print_mode == 'last':
        #Print_Step(t, belief_vec_temp, pol_temp)
        Plot_Belief_Polarization(belief_vec_temp, color='b', step=t, pol_value=pol_temp)
        
    ## Returns final value.
    return pol_hist
```
